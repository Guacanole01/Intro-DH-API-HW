```python
import requests
import time
import csv
```

```python
# Args: State's name (state_name - str), year it was founded (year - int),
#       lists of state newspapers (union - list), page number from NDPH page
#       (page - int)
# Use:  Given the state and year it was founded, it will find every newspaper
#       titled w its name and started before the state joined the US.
#       It will then return the list of newspapers

def find_newspapers(state_name, year, union, page):
    process_name = state_name.replace(" ", "+")

    url = f"https://chroniclingamerica.loc.gov/search/titles/results/?terms={process_name}&page={page}&format=json"
    info = requests.get(url)

    if info.status_code != 200:
        # raise ValueError(f"Request failed with status code {info.status_code}: {info.text}")

        if info.status_code == 429:
            print("Rate limit exceeded, waiting for 60 seconds")
            time.sleep(60)
            info = requests.get(url)
        else:
            raise ValueError(f"Request failed with status code {info.status_code}: {info.text}")

    state_info = info.json()

    for item in state_info['items']:
        if item['start_year'] < year:
            union.append(item)
    if state_info['endIndex'] == state_info['totalItems']:
        return union
    print(f"current page: {page} and current index: {state_info['endIndex']}")

    time.sleep(10)

    return find_newspapers(state_name, year, union, page + 1)
```

```python
# Args: List of states to search for (selected_states - list)
# Use:  Takes the states that the program will search through Chroicaling
#       America for and returns a dictionary where the key are those states and
#       the value is a nested list of the newspapers.

def find_states(selected_states):
    american_newspapers = {}
    for state, year in selected_states.items():
        newspaper_list = []
        print(f"the state name is {state} and the year is {year}")
        american_newspapers[state] = find_newspapers(state, year - 1, newspaper_list, 1)
        time.sleep(10)
    return american_newspapers
```
